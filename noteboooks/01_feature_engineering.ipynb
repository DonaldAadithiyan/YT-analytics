```python
import pandas as pd 
import os
import numpy as np

```

```python
inital_dataset = pd.read_csv("../data/raw/YouTube_Video_Metrics.csv" , engine="python",           # Use Python engine, more tolerant
    quoting=csv.QUOTE_ALL,
    on_bad_lines='skip', )
inital_dataset
```

```python
cleaned_dataset = inital_dataset.drop(columns=["title", "description"])
cleaned_dataset
```

```python
# Convert published_at to datetime
inital_dataset["published_at"] = pd.to_datetime(inital_dataset["published_at"])

# Create new columns
inital_dataset["year"] = inital_dataset["published_at"].dt.year
inital_dataset["month"] = inital_dataset["published_at"].dt.month
inital_dataset["day_of_week"] = inital_dataset["published_at"].dt.day_name()
inital_dataset["published_time"] = inital_dataset["published_at"].dt.time  


# Drop the original published_at column
inital_dataset = inital_dataset.drop(columns=["published_at"])

# Optionally, drop other columns as before
cleaned_dataset = inital_dataset.drop(columns=["title", "description"])
```

```python
cleaned_dataset = cleaned_dataset.rename(columns={"year": "published_year"})
cleaned_dataset = cleaned_dataset.rename(columns={"month": "published_month"})
cleaned_dataset = cleaned_dataset.rename(columns={"day_of_week": "published_day_of_week"})
```

```python
cleaned_dataset
```

```python
cleaned_dataset.columns
```

```python
cleaned_dataset["view_count_difference"] = cleaned_dataset["view_count_final"] - cleaned_dataset["view_count_initial"]
cleaned_dataset["like_count_difference"] = cleaned_dataset["like_count_final"] - cleaned_dataset["like_count_initial"]
```

```python
# Average view count difference per category
avg_view_diff = cleaned_dataset.groupby("category_id")["view_count_difference"].transform("mean")
cleaned_dataset["avg_view_diff_per_category"] = avg_view_diff

# Average likes difference per category
avg_likes_diff = cleaned_dataset.groupby("category_id")["like_count_difference"].transform("mean")
cleaned_dataset["avg_likes_diff_per_category"] = avg_likes_diff
```

```python
cleaned_dataset
```

```python
cleaned_dataset
```

```python
cleaned_dataset['channel_authority'] = np.log1p(cleaned_dataset['c_subscriber_count_initial'])
```

```python
cleaned_dataset['channel_avg_views'] = cleaned_dataset['c_view_count_initial'] / (cleaned_dataset['c_subscriber_count_initial'] + 1)
cleaned_dataset['channel_growth_potential'] = cleaned_dataset['channel_authority'] * cleaned_dataset['channel_avg_views']
cleaned_dataset['channel_virality_score'] = cleaned_dataset['channel_authority'] * np.log1p(cleaned_dataset['c_subscriber_count_initial'])
cleaned_dataset['log_channel_subs'] = np.log1p(cleaned_dataset['c_subscriber_count_initial'])
```

```python
cleaned_dataset['relative_views_to_category'] = cleaned_dataset['c_view_count_initial'] / (cleaned_dataset['avg_view_diff_per_category'] + 1)
cleaned_dataset['relative_likes_to_category'] = cleaned_dataset['channel_authority'] / (cleaned_dataset['avg_likes_diff_per_category'] + 1)
```
